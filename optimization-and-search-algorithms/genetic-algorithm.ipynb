```python
import random

random.seed(42)

genes = 'abcdefghijklmnopqrstuvwxyzABCDEFGHIJKLMNOPQRSTUVWXYZ1234567890!@#$%^&* ()_+-=[]{}|;:,.<>?'
target = 'Hello, World!'


class GeneticAlgorithm:
    def __init__(self, genes, target, mutation_rate, population_size):
        self.genes = genes
        self.target = target
        self.mutation_rate = mutation_rate
        self.population_size = population_size
        self.population = self.generate_population()


    def generate_population(self):
        generated_pop = [''.join(random.choice(self.genes) for _ in range(len(self.target))) for _ in range(self.population_size)]

        return  generated_pop

    def fitness(self, individual):
        return sum(1 for i, j in zip(individual, self.target) if i == j)

    def mutate(self, individual):
        return ''.join(random.choice(self.genes) if random.random() < self.mutation_rate else i for i in individual)
    
    def crossover(self, parent1, parent2):
        index = random.randint(0, len(self.target))
        return parent1[:index] + parent2[index:]
    
    def evolve(self):
        self.population = sorted(self.population, key=self.fitness, reverse=True)
        new_population = self.population[:self.population_size // 2]
        while len(new_population) < self.population_size:
            parent1 = random.choice(self.population)
            parent2 = random.choice(self.population)
            child = self.crossover(parent1, parent2)
            child = self.mutate(child)
            new_population.append(child)
        self.population = new_population
    
    def run(self):
        generation = 0
        while True:
            self.evolve()
            best = self.population[0]
            print(f'Generation: {generation}, Best: {best}, Fitness: {self.fitness(best)}')
            if best == self.target:
                break
            generation += 1


if __name__ == '__main__':
    ga = GeneticAlgorithm(genes, target, 0.1, 1000)
    ga.run()

```

**Flow of the Genetic Algorithm**
1. Generate the first generation from the gene pool
2. Sort this generation via the fitness function and take best fitted half
3. Crossover random parents to obtain offsprings until the size of the population
4. Mutate offsprings if mutate function is satisfied
5. If the target is reached, stop the program else generate a new generation and return to step 2

```python
genes = '1234567890'

target = '19051881'

random.seed(42)

ga = GeneticAlgorithm(genes, target, 0.1, 1000)
ga.run()
```
